```python
def move_file(source,destination,str_glob):
    allfiles = glob.glob(os.path.join(source, str_glob), recursive=True)
    if len(allfiles) < 1:
        sys.exit("Don't find file")
    # iterate on all files to move them to destination folder
    for file_path in allfiles:
        dst_path = os.path.join(destination, os.path.basename(file_path))
        shutil.move(file_path, dst_path)
```

```python
def consecutive(data, stepsize=1): #return 1 list of sub-arrays consecutive
    return np.split(data, np.where(np.diff(data) != stepsize)[0]+1)
```

```python
def process_hour(a):
    i=0
    while i < len(a):
        start= str(a[i][0])
        end=str(int(a[i][len(a[i])-1])+1)
        
        if len(start)==1:
            start="0"+start+":00"
        elif len(start)>1:
            if int(start)>=24:
                start=str(int(start)-24)
                if len(start)==1:
                    start= "0"+start+":00"
                else:
                    start=start+":00"
            else:
                start=start+":00"
        
        if len(end)==1:
            end="0"+str(int(end))+":00"
        elif len(end)>1:
            if int(end) >=24:
                end=str(int(end)-24)
                if len(end)==1:
                    end= "0"+end+":00"
                else:
                    end=end+":00"
            else:
                end=str(int(end))+":00"
        #time_period=time_period + start + "-" + end + ";"
        a[i]= start + "-" + end
        i+=1
    interval = ";".join([ i for i in a])
    return interval
```
